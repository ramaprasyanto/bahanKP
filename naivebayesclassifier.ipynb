```python
import csv
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', 1000)
df =pd.read_csv('data/baru/clear_text_testing_UPDATE.csv', delimiter=',', header=None, names=['text','lable'])

df
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(df['text'], 
                                                    df['lable'], 
                                                    random_state=0)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Fit the TfidfVectorizer to the training data specifiying a minimum document frequency of 5
vect = TfidfVectorizer(min_df=5, ngram_range=(1,2)).fit(X_train)
len(vect.get_feature_names())
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import roc_auc_score

X_train_vectorized = vect.transform(X_train)

model = MultinomialNB()
model.fit(X_train_vectorized, y_train)

predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
```

```python

feature_names = np.array(vect.get_feature_names())

sorted_tfidf_index = X_train_vectorized.max(0).toarray()[0].argsort()

print('Smallest tfidf:\n{}\n'.format(feature_names[sorted_tfidf_index[:10]]))
print('Largest tfidf: \n{}'.format(feature_names[sorted_tfidf_index[:-11:-1]]))
```

```python
sorted_coef_index = model.coef_[0].argsort()

print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```

```python
import csv

cols=['text']
tr = pd.read_csv('data/bersih/clear2.csv',delimiter=';',header=None,names=cols)
def testing_result(text):
    result_label = model.predict(vect.transform([text]))
    return result_label[0]

tr['new_label'] = tr['text'].apply(lambda t: testing_result(t))
tr
```

```python
#saving
with open('data/bersih/hasillogistic.csv', 'a') as f:
    tr.to_csv(f, header=None, columns = ["text","new_label"],index=False)
print("berhasil menyimpan")
```
