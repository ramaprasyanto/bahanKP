```python
from nltk.tokenize import WordPunctTokenizer
import re
from bs4 import BeautifulSoup

tok = WordPunctTokenizer()
pat1 = r'@[A-Za-z0-9_]+'
pat2 = r'https?://[A-Za-z0-9./]+'
pat3 = r'#[A-Za-z0-9]+'
pat4 = r'\[A-Za-z0-9]+'

combined_pat = r'|'.join((pat1, pat2, pat3,pat4))
#combined_pat = r'|'.join((combined, pat3))

def tweet_cleaner(text):
    soup = BeautifulSoup(text, 'lxml')
    souped = soup.get_text()
    stripped = re.sub(combined_pat, '', souped)
    try:
        clean = stripped.decode("utf-8-sig").replace(u"\ufffd", "?") 
    except:
        clean = stripped
    letters_only = re.sub("[^a-zA-Z0-9]", " ", clean)
    lower_case = letters_only.lower()
    words = tok.tokenize(lower_case)
    parsedData = [s.replace('b ','', 1) for s in words]
    parsedData1 = [s.replace('\nb', '') for s in parsedData]
    return (" ".join(parsedData1)).strip()


testing = df.text[:10000]
test_result = []
for t in testing:
    test_result.append(tweet_cleaner(t))
    
test_result
```
